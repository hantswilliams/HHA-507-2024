```python
import geopandas as gpd
import matplotlib.pyplot as plt
```

```python
# https://catalog.data.gov/dataset/national-obesity-by-state-d765a

file = 'https://raw.githubusercontent.com/hantswilliams/HHA-507-2024/refs/heads/main/Module10/geojson/National_Obesity_By_State.geojson'

df = gpd.read_file(file)

df.sort_values(by='Obesity', ascending=False)
```

```python
## visualize the data, so we can see the obesity rates by state,
## the geometry column is the shape of the state, and the Obesity column is the obesity rate
## using matplotlib, we can plot the data

## plot heatmap
#

df.plot(column='Obesity', cmap='coolwarm', linewidth=0.8, edgecolor='0.8')
plt.show()
```
